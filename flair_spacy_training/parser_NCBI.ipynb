```python
from bs4 import BeautifulSoup
def corpus_data(file):
    tokens = []
    categories_classes = []
    infile = open(file,"r")
    contents = infile.read()
    contents=contents.replace("<category=", "<category type =")
    soup = BeautifulSoup(contents,"lxml")
    categories = soup.find_all('category')
    
    for category in categories:
        tokens.append(category.text)
   
    for cat in categories:
        categories_classes.append(cat.attrs['type'])
        
    corpus = list(zip(tokens, categories_classes))
    return corpus
```

```python
corpus_training = corpus_data("NCBI_corpus/NCBI_corpus_training.txt")
corpus_testing = corpus_data("NCBI_corpus/NCBI_corpus_testing.txt")
corpus_training
```
